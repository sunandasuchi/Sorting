## **Sorted Array Implementation with step by step Output**

#### **Time Complexity: O(n log n)**

It has total **log n height** and at each height **n** amount of work is done. 

So, Total complexity is **nlogn** 

#### **Space Complexity: O(n)**

n = Total length of array

We are creating 2 array. 1 for Left and 1 for Right. Total space needed is **n = Left + Right**. 


```python
def mergeSort1(arr):
  if len(arr) >1:
    mid = len(arr) //2
    print("mid", arr[mid])
    L= arr[:mid]
    print("L=", L)
    R= arr[mid:]
    print("R=",R)
    mergeSort1(L)
    mergeSort1(R)

    res = merging(arr, L, R)
    print("Res: ", res)
  else:
    return arr
  return arr
def merging(arr, L, R):
    print("\n Inside Merging Function")
    print("arr: ", arr)
    print("Left", L)
    print("Right", R)
    i = 0
    j = 0
    k= 0

    while i<len(L) and j < len(R):
      if L[i] <= R[j]:
        arr[k] = L[i]
        print("\nIn first loop IF where L < R", arr)
        i+= 1
        k+= 1
      else:
        
        arr[k] = R[j]
        print("\nIn first loop ELSE where R < L", arr)
        j+= 1
        k+= 1
    while i <len(L):
      
      arr[k] = L[i]
      print("\nIn while Checking for items in L", arr)
      i += 1
      k += 1

    while j <len(R):
      
      arr[k] = R[j]
      print("\nIn while Checking for items in R", arr)
      j += 1
      k += 1
    
    print("Array at the End:",arr)
    return arr

    
```

```python
arr = [17,87,6,22, 41,3, 13,54]
ans = mergeSort1(arr)
print("\n\n\nSorted Array: ", ans)
```
